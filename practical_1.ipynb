```python
df['CIBIL_Score'] = df['CIBIL_Score'].astype(int)
df['Annual_Income'] = df['Annual_Income'].astype(float) #convert data types
```

```python
# normalization (min-max scalling)
df['Income_norm'] = (df['Annual_Income'] - df['Annual_Income'].min()) / (df['Annual_Income'].max() - df['Annual_Income'].min())

df['LoanAmount_norm'] = (df['Loan_Amount'] - df['Loan_Amount'].min()) / (df['Loan_Amount'].max() - df['Loan_Amount'].min())
```
