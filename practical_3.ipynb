```python
#Libraries
import pandas as pd
import numpy as np
```

```python
#load dataset
df = pd.read_csv("bank_loan_dataset.csv")
print(df.head())
```

```python
#create income group
df['Income_Group'] = pd.cut(df['Annual_Income'],
                           bins=[0, 300000, 700000, 1500000],
                           labels=['Low', 'Medium', 'High'])
```

```python
#group by income and statics
grouped = df.groupby('Income_Group')['Loan_Amount']

print("Mean:\n", grouped.mean())
print("Median:\n", grouped.median())
print("Minimum:\n", grouped.min())
print("Maximum:\n", grouped.max())
print("Standard Deviation:\n", grouped.std())
```

```python
#create list of each category
low_income = df[df['Income_Group'] == 'Low']['Loan_Amount'].tolist()
medium_income = df[df['Income_Group'] == 'Medium']['Loan_Amount'].tolist()
high_income = df[df['Income_Group'] == 'High']['Loan_Amount'].tolist()

print("Low Income Loan List:", low_income[:10])
print("Medium Income Loan List:", medium_income[:10])
print("High Income Loan List:", high_income[:10])
```

```python
#group by approval
approved = df[df['Loan_Approved'] == 1]
not_approved = df[df['Loan_Approved'] == 0]
```

```python
#function for statics
def stats(data, name):
    print("\nStatistics for", name)
    print("Mean:\n", data.mean())
    print("Standard Deviation:\n", data.std())
    print("Percentiles:\n", data.quantile([0.25, 0.5, 0.75]))
```

```python
#apply function
stats(approved.select_dtypes(include=np.number), "Approved Loans")
stats(not_approved.select_dtypes(include=np.number), "Not Approved Loans")
```
